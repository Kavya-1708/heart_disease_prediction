# Logistic Regression

```python
# import neccessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import classification_report
```

```python
# Split data into features and target
X = df.drop('target', axis=1)
y = df['target']
```

```python
#Split the dataset (70:30 ratio)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Build and train the model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred = model.predict(X_test)
y_pred
```

```python
# Evaluate the model
conf_matrix = confusion_matrix(y_test, y_pred)
conf_matrix
```

```python
# Plot the confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Disease', 'Disease'], yticklabels=['No Disease', 'Disease'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# classification report
print(classification_report(y_test,y_pred))
```

```python
# Accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

# Decision Tree Classifier

```python
# Import neccessary libraries
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import classification_report
import graphviz
from sklearn import tree
```

```python
# Prepare the data
X = df.drop('target', axis=1)
y = df['target']
```

```python
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Build decision tree
dt_model = DecisionTreeClassifier(random_state=42, max_depth=4)
dt_model.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred_dt = dt_model.predict(X_test)
```

```python
# Evaluate the model
cm_dt = confusion_matrix(y_test, y_pred_dt)
cm_dt
```

```python
# Plot confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Greens',xticklabels=['No Disease', 'Disease'],yticklabels=['No Disease', 'Disease'])
plt.title('Decision Tree - Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```

```python
# Classification report
print(classification_report(y_test, y_pred_dt))
```

```python
# Visualize the decision tree using graphviz package
dot_data = tree.export_graphviz(dt_model, out_file=None,
                                feature_names=X.columns,
                                class_names=['No Disease', 'Disease'],
                                filled=True, rounded=True,
                                special_characters=True)

# Render the tree
graph = graphviz.Source(dot_data)
graph.render("decision_tree")  # Saves as 'decision_tree.pdf'
graph.view()
```

```python
# accuracy of the model
accuracy_dt = accuracy_score(y_test, y_pred_dt)
accuracy_dt
```

# Random forest classifier

```python
# import neccessary libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import classification_report
from sklearn import tree
import graphviz
```

```python
# Separate features and target
X = df.drop('target', axis=1)
y = df['target']
```

```python
# Divide the dataset in 70:30 ratio
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Build the model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Predict
y_pred = rf_model.predict(X_test)
y_pred
```

```python
# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Plot confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Oranges',xticklabels=['No Disease', 'Disease'], yticklabels=['No Disease', 'Disease'])
plt.title('Random Forest - Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```

```python
# classification report
print(classification_report(y_test, y_pred))
```

```python
# Accuracy
acc = accuracy_score(y_test, y_pred)
acc
```

# Model which is best

```python
# Predict which model is best
models = ['Logistic Regression', 'Decision Tree', 'Random Forest']
best_model = models[[accuracy, accuracy_dt, acc].index(max(accuracy, accuracy_dt, acc))]
print(" Best Performing Model:", best_model)
```
